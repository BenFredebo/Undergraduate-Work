```python
# Notebook for Assignment # 1:

# imports modules:
import numpy as np
import matplotlib.pyplot as plt
```

```python
# question # 1 a):

# first define the function in complex representation:

def ellipse(phi, ar=1, ai=1, br=1, bi=1):
    '''
    Takes an angle input as well as 
    the real and imaginary components of
    the coefficients and returns x and y
    coordinates to plot
    '''
    z = (ar+ai*1j) * np.exp(phi*1j ) + (1j*bi+br) * np.exp(-phi*1j)
    
    return z

# creates list of N angles making up the ellipse 
# from Phi = 0, Phi = 2pi

N = 500
phis = np.linspace(0, 2*np.pi, N)
    
# Calculates the complex values in rectangular representation
# for different values of a, b: 

a_r = 5; a_i = 12; b_r = 3; b_i = 4 

ZZ = ellipse(phis, a_r, a_i, b_r, b_i)

# calculates the semi-major and minor axes as well as the phase
# and prints them to the output:

semi_minor = np.abs(np.sqrt(a_r**2 +a_i**2) - np.sqrt(b_r**2 +b_i**2))
semi_major = np.sqrt(a_r**2 +a_i**2) + np.sqrt(b_r**2 +b_i**2)

phase = (np.arctan(a_i/a_r) + np.arctan(b_i/b_r))/2

print(f"For ar = {a_r}, ai = {a_i}, br = {b_r}, and bi = {b_i}")
print()
print(f"The semi-major axis is: {semi_major}")
print(f"The semi-minor axis is: {semi_minor}")
print()
print(f"The Phase is: {phase}")
print()
      
      
# splits into real and imaginary components:

XX = [num.real for num in ZZ]
YY = [num.imag for num in ZZ]

# plots the ellipse:

ff, (ax1) = plt.subplots(1, 1)
ff.set_figwidth(6)
ff.set_figheight(4)

ax1.plot(XX,YY, color='#1a2081')#, label=r"$H_4(\omega)$")
ax1.set_title("Ellipse in Complex plane")
ax1.set_xlabel(r"Re[z]")
ax1.set_ylabel(r"Im[z]" )
#ax1.legend()
ax1.grid(True, alpha=0.3)

```

We can then use the above plot to check the following work with different values of Re(a), Im(a), Re(b), Im(b). It is easy to see that the equation involved produces an ellipse. We will show this more rigorously in the following analysis.

To find the axes and the orientation of the ellipse and to show that it is indeed an ellipse, we start with the initial definition:

$$z = ae^{i\phi} + be^{−i\phi}$$

Now, since $a,b \in \mathbb{c}$, redefine them as:

$$ a = |a| e^{i\theta_a} \ ; \ b = |b| e^{-i\theta_b}$$ 

Thus:

$$ z = |a| e^{i\theta_a}e^{i\phi} + |b| e^{-i\theta_b}e^{−i\phi}$$

$$\implies  \ \ \ \ \ \  z = |a| e^{i(\theta_a+\phi)} + |b| e^{-i(\theta_b+\phi)}$$

To produce an overall phase for the above equation, we may define the following quantities in terms of the angles associated with a,b:

$$ \xi = \frac{\theta_a + \theta_b}{2} \ ; \ \eta = \frac{\theta_a - \theta_b}{2}$$

To use the sum and difference relations, we may add and subtract half of the missing quantitieas:

$$z = |a| e^{i(\theta_a+\phi +\frac{\theta_b}{2}-\frac{\theta_b}{2})} + |b| e^{-i(\theta_b+\phi+\frac{\theta_a}{2}-\frac{\theta_a}{2})}$$

Plugging in $\xi$ and $\eta$ gives us:

$$ z = |a| e^{i(\xi+\phi +\eta)} + |b| e^{-i(\xi+\phi -\eta)}$$

$$ \implies \ \ \ \ \ \ e^{i\xi} \left( |a| e^{i(\phi - \eta)} + |b| e^{-i(\phi - \eta)} \right)$$

We can let $\psi = \phi - \eta$:

$$ z =e^{i\xi} \left( |a| e^{i\psi} + |b| e^{-i\psi} \right)$$

Now both terms in the above can be written in another standard complex representation such that $e^{i\psi} = cos\psi + isin\psi$:

$$ z = e^{i\xi} \left( |a|(cos\psi +isin\psi) + |b| (cos\psi-isin\psi) \right)$$

Seperating into the real and imaginary components:

$$ z = e^{i\xi} \left[(|a|+|b|)cos\psi + i(||a| - |b||)sin\psi) \right]$$

Now we have that:

$$ \Re{(z)} = e^{i\xi} (|a|+|b|)cos\psi \ ; \ \Im{(z)}=e^{i\xi} (||a| - |b||)sin\psi$$

If we think of a parametric curve in the complex plane, the above equations describe and ellipse perfectly. The cosine and sine terms in the real and imaginary components respectively will always be $\frac{\pi}{2}$ units out of phase producing an ellipse. The half-axes of the ellipse are the coefficients of the cosine and sine functions such that:

$$ r_{major} = |a| + |b| \ ; \ r_{minor} = ||a| - |b|| $$

We can also say that the relative orientation is governed by the phase, with the phase factor:

$$ phase = \xi = \frac{\theta_a + \theta_b}{2} = \frac{arctan\left(\frac{a_i}{a_r}\right) + arctan\left(\frac{b_i}{b_r}\right)}{2}$$

### Question 2, part 2:

*(1pt) Determine the (Taylor or Laurent) series for the function 
$f(z)=\frac{cosz}{z-1}$ about z = 1.*

In order to expand this function around $z=1$, we first must investigate it's properties. The numerator is an entire function and can be easily converted into a series. The denominator, however, is a bit more complicated as it $=0$ at $z=1$.

We can first start with the Taylor series of $cos z$. As:

$$cosz = \sum^{\infty}_{n=0} \frac{f^{(n)}(a)}{n!} (z-a)^n$$

This is fairly easy to find, with $a=1$ and the normal cyclical derivatives of sine and cosine. The first few terms can be seen as follows:

$$ cosz = cos(1) - sin(1) (z-1) - \frac{1}{2}cos(1)(z-1)^2 + \frac{1}{6} sin(1) (z-1)^3 + \frac{1}{24}cos(1) (z-1)^4 ...$$

We can combine this with the denominator into a Laurent series quite simply, while noting that we have a simple pole at $z=1$ and so the function will not converge for that value. The expansion is then:

$$ f(z) = \frac{cosz}{z-1} = \frac{cos(1)}{z-1} - sin(1) - \frac{1}{2}cos(1)(z-1) + \frac{1}{6} sin(1) (z-1)^2... $$

Or more compactly:

$$ f(z) = \sum^{\infty}_{n=0} \frac{d^{n}}{dx^{n}} cos(z)\bigg|_{z=1} \frac{(z-1)^{n-1}}{n!}$$

### Question 2, Part 3:

*(2pt) Determine all Taylor or Laurent series of the function $\frac{1}{z^2+1}$ about z = i.*

The first step in determining the Series expansions of this function is to factor it:

$$ f(z) = \frac{1}{z^2+1} = \frac{1}{(z-i)(z+i)}$$

Further simplifying gives:

$$f(z) = \frac{i}{2} \left(\frac{1}{z+i}-\frac{1}{z-i}\right)$$

It is clear that we have more than one region that we need to expand around. This is due to having two poles at $z=\pm i$. Our first region of consideration is expanded around $z=i$ until we reach $z=-i$. Thus our region of convergence can be found as follows:

$$R < |z-a| \longrightarrow R=2<|z-i|$$

Our second region is anything outside of this circle such that:

$$R > |z-a| \longrightarrow R=2 > |z-i|$$

$$\frac{1}{z+i} = \frac{1}{z-i+i+i} = \frac{1}{(z-i) +2i} = \frac{1}{1+\frac{z-i}{2i}}\frac{1}{2i}$$

This is roughly in a form where we can use the geometric series:

$$  \frac{1}{1+s} = \sum^{\infty}_{n=0} (-1)^n s^n \ for \ |s| < 1 $$

### 3. Waves

*(3pt) The famous double-slit experiment illustrates the wave nature of light, as well as the wave nature of the probability amplitude in Quantum Mechanics. A plane
wave of wavelength λ is incident on a plate with narrow parallel slits which are a
distance d apart, and the interference pattern is observed on the screen behind the
plate. The slits are located symmetrically about z = 0. The distance between the
plate and the screen is L ≫ d. Represent the wave coming out of each of the slits
as a complex exponential, and find the intensity distribution I(z) on the screen in
terms of I(z=0) for a) three slits, b) N slits. Compute the intensity by taking the
absolute value squared of the amplitude. Ignore the diffraction effects due to the
finite width of the slits. Make a plot of I(z).*


To begin out analysis, the properties of the system must be taken into account. We have that the plane wave is incident with a wavelength $\lambda$ which will contribute to the path difference. Further, the slits are orientated symmetrically about a central point $z=0$ with seperations of $d$. The distance to the screen is $L$. Since this is much greater than the slit seperation, we can approximate the output waves from the slits as parallel plane waves which will aid in the analysis. We have N slits that will all have a plane wave output with amplitudes and phases respectively.

The path difference between consecutive slits is a function of the wavelength, distance between the slits, and the angle on the screen of interest. By extending rays from the slit openings to the screen between consecutive slits, a right triangle appears with principle angle, $\phi$, that defines the angular height along the screen. Between, lets say, plane wave 1 and 2, we have that the path difference is given by:

$$ \delta_{path}^{12} = \frac{2\pi}{\lambda} d sin\phi$$

Where d is the slit seperation and $\lambda$ is the wavelength of the light. We can take this a step further and extend it to the path difference between higher numbered slits as:

$$ \delta_{path}^{1n} = \frac{2\pi}{\lambda} (d+x) sin\phi$$

Where x is the seperation between the first slit and the slit of interest. Since the slits are evenly seperated, we can say that this x value is an integer multiple of d such that:

$$ \delta_{path}^{1n} = \frac{2\pi}{\lambda} (N\cdot d) sin\phi$$

With $N=n-1$ for $n=1,2,3...$.

We can further manipulate this by assuming that the angle involved is small. This means that $\phi \approx \frac{z}{L}$, where $z$ is the height of interest up the screen and $L$ is the distance to the screen. So:


$$ \delta_{path}^{1n} = \frac{2\pi}{\lambda} \frac{Ndz}{L} \ for \ N = 0,1,2...$$

We can also say that the path difference for an arbitrary n is a multiple of the path difference of the first two slits as:

$$ \delta_{path}^{1n} = \delta ^{11}_{path} \cdot N $$

For simplicity, let $\delta_{path}^{11} = \delta$.

We now move to the waves describing the oncoming light itself. A plane wave has the form:

$$E = E_0 e^{i(kx-\omega t+ \psi)}$$

For subsequent waves, with the accrued path difference from the slits:

$$E_1 = E_0 e^{i(kx-\omega t+ \psi)}$$

$$E_2 = E_0 e^{i(kx-\omega t+ \psi + \delta)}$$

$$E_3 = E_0 e^{i(kx-\omega t+ \psi + 2\delta)}$$

etc...

Factoring out the common wave if we assume all field magnitudes are equal:

$$E_2 = E_0 e^{i(kx-\omega t+ \psi )}e^{i\delta}$$

$$E_3 = E_0 e^{i(kx-\omega t+ \psi)} e^{2\delta}$$

etc...

We can then say that:

$$E_2 = E_1e^{i\delta}$$

$$E_3 = E_1e^{2\delta}$$

etc...

And: 

$$E_N = E_1e^{iN\delta}$$


Our electric field at the screen is a combination of all of the waves from the slits. This can be written as:

$$E_T = \sum^{N}_{n=1} E_1 e^{i\delta N-1}$$

This looks like a finite geometric series as:

$$ Sum = \sum^{N}_{n=1} a r^{n-1} = a\frac{1-r^n}{1-r}$$

If we let $a=E_1$ and $r=e$ we arrive at an expression:

$$ E_T = E_1 \left( \frac{1-e^{iN\delta}}{1-e^{i\delta}}\right)$$


Since we are ignoring the diffraction effects, we can say that $|E_1|=const.$ in $z$. In order to get the Interference pattern intensity, we need to take the modulus as:

$$ I(z) = |E_t|^2 = |E_1|^2 \left( \frac{1-e^{iN\delta}}{1-e^{i\delta}}\right)\left( \frac{1-e^{-iN\delta}}{1-e^{-i\delta}}\right)$$

We may now deal with the numerator and denominator seperately. Firstly, the denominator:

$$ (1-e^{-i\delta})\cdot (1-e^{i\delta}) = 2 - e^{i\delta} -e^{-i\delta} = 2-(e^{i\delta} +e^{-i\delta} = 2-2cos(\delta) = 2(1-cos(\delta))$$

Further simplifying:

$$ 2(1-cos(\delta)) = 4 sin^2\left(\frac{\delta}{2}\right)$$

Now for the numerator. We can do the exact same simplification as above but with $\delta \rightarrow N\delta$:

$$ \left( 1-e^{iN\delta} \right) \left( 1-e^{-iN\delta} \right) = 4 sin^2\left(\frac{N\delta}{2}\right)$$

Our whole expression then becomes:

$$ I(z) = |E_t|^2 = |E_1|^2 \frac{4 sin^2\left(\frac{N\delta}{2}\right)}{4 sin^2\left(\frac{\delta}{2}\right)} = |E_1|^2\frac{ sin^2\left(\frac{N\delta}{2}\right)}{ sin^2\left(\frac{\delta}{2}\right)}$$

With $\delta = \frac{2\pi}{\lambda} \frac{zd}{L}$. We can now plot this for a number of values of N.

```python
# defines the function to be plotted:

def nslit_interference(z, N, mod_E = 1, lamb =700*10**(-9), d=10**(-6),L = 1 ):
    '''
    takes in list of z values to plot
    as well as the number of slits,N,
    and returns the intensity at the
    particular z value
    
    optional:
    Value for electric field modulus 
    mod_E. Defaults = 1
    
    Wavelength defaults to red light
    lamb = 700 nm
    
    width of slit defaults to 1 micrometre
    
    Distance to screen defaults to 1 metre
    '''
    
    delta = ((2*np.pi*d)/(lamb*L))*z
    
    I = mod_E * (((np.sin((N*delta)/2))**2)/((np.sin((delta)/2))**2))
    
    return I
```

```python
# calculates the intensity as a function of z for z E [0,1] m:

ZI = np.linspace(-1,1,500)

# creates list of lists for plotting for 1<N<5 N values:

II = [nslit_interference(ZI, N+2) for N in range(10)]
```

```python
# plots the figure:

ff, (ax1) = plt.subplots(1, 1)
ff.set_figwidth(8)
ff.set_figheight(6)

# colours and labels:

cols = ['firebrick', 'salmon', 'darkorange', 'gold', 'seagreen', 
        'lightseagreen', 'darkcyan', 'cornflowerblue', 'navy', 'indigo']
labs = [f'N={1+i}' for i in range(10)]

# nomarlized the intensity to compare:
# plots the patterns
for i in range(len(II)):
    ax1.plot(ZI, II[i]/(max(II[i])), color=cols[-i], label=labs[i], alpha = 1-i*0.07,zorder=25-2*i)
ax1.set_title("Intensity pattern for Multiple Slits, from N=1 to N=10:")
ax1.set_xlabel(r"Height, z, Along Screen [m]")
ax1.set_ylabel(r"Normalized Intensity Units [i.u.]" )
ax1.legend()
ax1.grid(True, alpha=0.3)
```
